    Use Q9_a.csv

```python
Cars=pd.read_csv("Q9_a.csv")
WT=pd.read_csv("Q9_b.csv")
```

```python
Cars
```

```python
WT
```

```python
sns.kdeplot(Cars["speed"],label="speed")
sns.kdeplot(Cars["dist"],label="dist")
sns.kdeplot(WT["SP"],label="SP")
sns.kdeplot(WT["WT"],label="WT")
plt.legend()
```

```python
from scipy.stats import skew,kurtosis
```

```python
print("SKEWNESS FOR CAR DATA:",skew(Cars.iloc[:,1:]))
print("==="*20)
print("Kurtosis FOR CAR DATA:",kurtosis(Cars.iloc[:,1:]))
```

```python
print("SKEWNESS FOR WT DATA:",skew(WT.iloc[:,1:]))
print("==="*20)
print("Kurtosis FOR WT DATA:",kurtosis(WT.iloc[:,1:]))
```
